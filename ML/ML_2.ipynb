```python
data.dropna(axis = 1)
```

splitting

```python
from sklearn.model_selection import train_test_split
```

```python
X = data.drop(["Prediction" , "Email No."] , axis = 1 )
Y = data["Prediction"]
```

```python
x_train , x_test , y_train , y_test = train_test_split( X , Y , random_state = 42 , test_size = 0.3)
```

Model training

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
km = KNeighborsClassifier()
km.fit(x_train , y_train)
y_train_pred = km.predict(x_train)
y_test_pred = km.predict(x_test)
```

```python
from sklearn.svm import SVC
```

```python
sm = SVC()
sm.fit(x_train , y_train)
y_train_pred_sm = km.predict(x_train)
y_test_pred_sm = km.predict(x_test)
```

Model evaluation

```python
from sklearn.metrics import confusion_matrix , classification_report
```

```python
print(confusion_matrix( y_test , y_test_pred) )
print(confusion_matrix( y_train , y_train_pred) )
```

```python
print(classification_report( y_test , y_test_pred) )
print(classification_report( y_train , y_train_pred) )
```

```python
print(confusion_matrix( y_test , y_test_pred_sm) )
print(confusion_matrix( y_train , y_train_pred_sm) )
```

```python
print(classification_report( y_test , y_test_pred_sm) )
print(classification_report( y_train , y_train_pred_sm) )
```
