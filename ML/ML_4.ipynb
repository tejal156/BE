```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
learning_rate = 0.1
starting_point  = 2
no_iterations = 50
ans = []

def fun(x):
  return (x+3)**2

def dfun(x):
  return 2 * (x+3)
```

```python
def algo( x , lr  , no_i):
  ans.append(x)
  for i in range(no_i):
    x = x - lr * dfun(x)
    ans.append(x)
    print( f"i: {i}, x:{x}, fun(x):{fun(x)}" )
  return x
```

```python
algo( starting_point, learning_rate, no_iterations )
```

```python
sns.scatterplot(ans)
```
